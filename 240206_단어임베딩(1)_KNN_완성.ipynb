# 단어 데이터 준비 및 정수 인코딩

```python
# 데이터분석 4종 세트
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_path = '/content/drive/MyDrive/Colab Notebooks/data/connects_lab/'
customer_df = pd.read_csv(base_path+'Data_1_TalkD_customer_240124_modi.csv', encoding='utf-8')
course_df = pd.read_csv(base_path+'Data_2_TalkD_course_240124_modi.csv',encoding='utf-8')
rating_df = pd.read_csv(base_path+'Data_3_ratings_df_240124_modi.csv',encoding='utf-8')
```

```python
# Extract and flatten the keywords from the 'Parenting_Concerns' and '관련 육아고민' columns
parenting_concerns = customer_df['Parenting_Concerns'].str.split(', ')
customer_words = [word.strip() for sublist in parenting_concerns for word in sublist]
course_concerns = course_df['관련 육아고민'].str.split(', ')
course_words = [word.strip() for sublist in course_concerns for word in sublist]

# Create a unique list of keywords
combined_words = list(set(customer_words + course_words))
```

```python
# Initialize and fit the tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(combined_words)

# Convert the texts to sequences and flatten to 1D array
sequences = tokenizer.texts_to_sequences(combined_words)
data = np.array([item for sublist in sequences for item in sublist])
```

```python
print(data)
```

# 임베딩 매트릭스 생성

```python
# Define the embedding layer
embedding_dim = 100
word_index = tokenizer.word_index
num_words = len(word_index) + 1
embeddings_index = {word: np.random.rand(embedding_dim) for word in word_index.keys()}  # Dummy embeddings
embedding_matrix = np.zeros((num_words, embedding_dim))
for word, i in word_index.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
```

# 모델 정의 및 컴파일

```python
# Define the model
model = Sequential()
model.add(Embedding(num_words, embedding_dim, input_length=1, weights=[embedding_matrix], trainable=False))
model.add(Dense(16, activation='relu'))
num_classes = 10  # Define your number of classes here
model.add(Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Reshape data for training and use the data itself as labels (modify as needed)
data = data.reshape(-1, 1)
labels = data
model.fit(data, labels, epochs=3, batch_size=32)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from tensorflow.keras.models import Model
from sklearn.decomposition import PCA
```

```python
# 'model'은 앞서 정의하고 학습한 Keras 모델입니다.
embedding_layer = model.layers[0]
weights = embedding_layer.get_weights()[0]
```

# KNN

```python
# 인덱스를 단어로 매핑 (Tokenizer의 인덱스를 그대로 사용)
index_to_word = {i: word for word, i in word_index.items()}
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import normalize

# 임베딩 가중치 정규화
normalized_weights = normalize(weights)

# 이웃의 수 정의
n_neighbors = 5  # 필요에 따라 이 값을 조정하세요

# KNN 모델 초기화 및 학습
knn = NearestNeighbors(n_neighbors=n_neighbors)
knn.fit(normalized_weights)

# 각 단어에 대한 가장 가까운 이웃 찾기
neighbors = knn.kneighbors(normalized_weights, return_distance=False)

# NearestNeighbors의 결과를 처리하고 인덱스를 맞춤
word_neighbors = {}
for i, neighbor in enumerate(neighbors):
    word = index_to_word.get(i + 1, 'N/A')
    if word != 'N/A':
        # 존재하는 단어에 대해서만 이웃 단어를 찾음
        word_neighbors[word] = [index_to_word.get(n + 1, 'N/A') for n in neighbor]

# 각 단어의 가장 가까운 이웃 출력
for word, neighbor in word_neighbors.items():
    print(f"단어: {word}, 이웃: {neighbor}")
```

```python
for word, neighbors in word_neighbors.items():
    print(f"단어: {word}")
    word_index = tokenizer.word_index[word] - 1  # 'word'의 인덱스
    for neighbor in neighbors:
        if neighbor != 'N/A':
            neighbor_index = tokenizer.word_index[neighbor] - 1  # 'neighbor'의 인덱스
            distance = np.linalg.norm(normalized_weights[word_index] - normalized_weights[neighbor_index])
            print(f"  이웃: {neighbor}, 거리: {distance:.2f}")
        else:
            print("  이웃: N/A, 거리: N/A")
```

# T-SNE 시각화

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# t-SNE 모델을 사용하여 임베딩을 2D 공간으로 축소
# perplexity 값을 데이터셋 크기의 약 절반으로 설정
tsne = TSNE(n_components=2, random_state=0, perplexity=len(normalized_weights)//2)
reduced_embeddings = tsne.fit_transform(normalized_weights)


# 단어 임베딩을 2D 공간에 플롯
plt.figure(figsize=(12, 12))
for i, word in enumerate(index_to_word.values()):
    plt.scatter(reduced_embeddings[i, 0], reduced_embeddings[i, 1])
    plt.annotate(word, xy=(reduced_embeddings[i, 0], reduced_embeddings[i, 1]), xytext=(5, 2), textcoords='offset points', ha='right', va='bottom')

# 각 단어에 대한 가장 가까운 이웃들의 위치를 시각화
for word, neighbors in word_neighbors.items():
    word_index = tokenizer.word_index[word] - 1
    for neighbor in neighbors:
        neighbor_index = tokenizer.word_index[neighbor] - 1
        plt.plot([reduced_embeddings[word_index, 0], reduced_embeddings[neighbor_index, 0]],
                 [reduced_embeddings[word_index, 1], reduced_embeddings[neighbor_index, 1]], 'r-')

plt.show()
```

```python
n_samples = weights.shape[0]
perplexity_value = min(40, n_samples - 1)  # perplexity는 n_samples보다 작아야 함

tsne = TSNE(n_components=2, verbose=1, perplexity=perplexity_value, n_iter=300)
tsne_results = tsne.fit_transform(weights)
```

```python
plt.figure(figsize=(10, 10))

# 무작위 오프셋을 위한 범위 설정
offset_range = 10  # 오프셋 범위를 조정할 수 있습니다.

for i, word in enumerate(word_index.keys()):
    if i >= 1000:  # 처음 1000개 단어만 표시
        break
    x, y = tsne_results[i, 0], tsne_results[i, 1]
    # 무작위 오프셋 추가
    x_offset = np.random.uniform(-offset_range, offset_range)
    y_offset = np.random.uniform(-offset_range, offset_range)

    plt.scatter(x, y)
    plt.annotate(word, xy=(x, y), xytext=(x_offset, y_offset),
                 textcoords='offset points', ha='right', va='bottom')

plt.show()
```
